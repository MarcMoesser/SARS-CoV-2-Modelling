```python
import os
```

```python
good_list = []

for file in os.listdir("./test_stuff"):
    name = file[:-20]
    good_list.append(name)
```

```python
good_list
```

```python
for file in os.listdir("./ligand_alignment/aligned"):
    name = file.split("_")[0]+"_"+file.split("_")[2][:-4]
    if name in good_list:
        os.system("mv ./ligand_alignment/aligned/"+file+" ./ligand_alignment/aligned_new")
```

```python
cov_frags = []
for file in os.listdir("./ligand_alignment/aligned_new/"):
    frag = file.split("_")[2][:-4]
    cov_frags.append(frag)
cov_frags = list(set(cov_frags))
for file in os.listdir("./ligand_alignment/fragments"):
    name = file.split("_")[0][5:]

    if name in cov_frags:
        os.system("cp ./ligand_alignment/fragments/"+file+" ./ligand_alignment/fragments_new")
```

```python
sub_mols = [i.split("_")[0] for i in good_list]
for file in os.listdir("./ligand_alignment/submissions"):
    name = file.split("_")[0]
    if name in sub_mols:
        os.system("cp ./ligand_alignment/submissions/"+file+" ./ligand_alignment/submissions_new")
```
